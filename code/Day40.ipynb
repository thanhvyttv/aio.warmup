```python
df["Sodium"].describe()
```

```python
sodium_mean = df["Sodium"].mean()
sodium_mean
```

```python
sodium_q3 = df["Sodium"].quantile(0.75)
sodium_q1 = df["Sodium"].quantile(0.25)
sodium_iqr = sodium_q3 - sodium_q1
sodium_iqr
```

```python
upper_limit = sodium_q3 + 1.5 * sodium_iqr
upper_limit
```

```python
# Find Outlier of Sodium and remove it
data = df[df["Sodium"] < upper_limit]
data
```

```python
group_category_item = (
    data.groupby(by="Category")["Item"].count().to_frame().reset_index()
)
group_category_item.rename(columns={"Item": "Item_counts"}, inplace=True)
group_category_item
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    data=group_category_item,
    x="Category",
    y="Item_counts",
    palette="tab10",
    hue="Category",
)
plt.title("Biểu đồ số lượng Items có trong Category")
plt.ylabel("Item Counts")
plt.xticks(rotation=30)
plt.show()
```
