```python
import matplotlib.pyplot as plt
import numpy as np

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temperatures = [22, 24, 25, 23, 26, 28, 27]

# fig = plt.figure(figsize=(8,6))

plt.plot(
    days, temperatures, color="red", linestyle="--", marker="o", label="Temperature"
)
plt.title("Weekly Temperature")
plt.xlabel("Day")
plt.ylabel("Temperature °C")
plt.legend()
plt.show()
```

### Sinh ngẫu nhiên một mảng 1000 số theo phân phối chuẩn (mean=0, std=1). Vẽ histogram với:
• 30 bins
• Màu xanh lá cây
• Đường viền màu đen

```python
import numpy as np

x = np.random.normal(0, 1, 1000)
plt.hist(x, bins=30, density=True, alpha=0.7, color="green", edgecolor="black")
plt.title("Histogram của 1000 số theo phân phối chuẩn")
plt.xlabel("Giá trị")
plt.ylabel("Tần số")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Tạo figure và các subplot
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Subplot 1: Vẽ đồ thị hàm sin
axs[0].plot(x, y1, color="blue")
axs[0].set_title("Đồ thị hàm sin(x)")

# Subplot 2: Vẽ đồ thị hàm cos
axs[1].plot(x, y2, color="red")
axs[1].set_title("Đồ thị hàm cos(x)")

# Điều chỉnh khoảng cách giữa các subplot
plt.tight_layout()

# Hiển thị figure
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu
quarters = ["Q1", "Q2", "Q3", "Q4"]
sales_2022 = [120, 135, 145, 160]
sales_2023 = [130, 148, 155, 170]

# Vị trí các cột trên trục x
x = np.arange(len(quarters))
width = 0.35  # Độ rộng của mỗi cột

# Tạo figure và axes
fig, ax = plt.subplots(figsize=(8, 6))

# Vẽ biểu đồ cột cho năm 2022
rects1 = ax.bar(x - width / 2, sales_2022, width, label="2022", color="skyblue")

# Vẽ biểu đồ cột cho năm 2023
rects2 = ax.bar(x + width / 2, sales_2023, width, label="2023", color="salmon")

# Thêm chú thích (legend)
ax.legend()

# Thêm tiêu đề
ax.set_title("Quarterly Sales Comparison")

# Thêm nhãn cho trục x và y
ax.set_xlabel("Quý")
ax.set_ylabel("Doanh thu")

# Đặt ticks và nhãn cho trục x
ax.set_xticks(x)
ax.set_xticklabels(quarters)


# Thêm nhãn giá trị trên mỗi cột (optional)
def autolabel(rects):
    """Gắn nhãn giá trị lên trên mỗi cột."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate(
            "{}".format(height),
            xy=(rect.get_x() + rect.get_width() / 2, height),
            xytext=(0, 3),  # 3 điểm offset theo chiều dọc
            textcoords="offset points",
            ha="center",
            va="bottom",
        )


autolabel(rects1)
autolabel(rects2)

# Điều chỉnh layout để tránh nhãn bị cắt
fig.tight_layout()

# Hiển thị biểu đồ
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ["A", "B", "C", "D"]
values1 = [20, 35, 30, 25]
values2 = [22, 38, 25, 28]

x = np.arange(len(categories))
width = 0.4

fig, ax = plt.subplots()
ax.bar(x - width / 2, values1, width, label="Bộ dữ liệu 1", alpha=0.7)
ax.bar(x + width / 2, values2, width, label="Bộ dữ liệu 2", alpha=0.7)

ax.set_ylabel("Giá trị")
ax.set_title("Nhiều bar tại cùng vị trí (ít khuyến khích)")
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ["A", "B", "C", "D"]
values1 = [20, 35, 30, 25]
values2 = [25, 32, 34, 20]
values3 = [30, 38, 28, 31]

fig, ax = plt.subplots()
ax.bar(categories, values1, label="Phần 1", color="skyblue")
ax.bar(categories, values2, bottom=values1, label="Phần 2", color="salmon")
ax.bar(
    categories,
    values3,
    bottom=np.array(values1) + np.array(values2),
    label="Phần 3",
    color="lightgreen",
)

ax.set_ylabel("Giá trị")
ax.set_title("Biểu đồ cột chồng")
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ["A", "B", "C", "D"]
values1 = [20, 35, 30, 25]
values2 = [25, 32, 34, 20]
values3 = [30, 38, 28, 31]

x = np.arange(len(categories))  # Vị trí cho các nhóm
width = 0.25  # Độ rộng của mỗi bar

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, values1, width, label="Nhóm 1")
rects2 = ax.bar(x, values2, width, label="Nhóm 2")
rects3 = ax.bar(x + width, values3, width, label="Nhóm 3")

ax.set_ylabel("Giá trị")
ax.set_title("Biểu đồ cột nhóm")
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

fig.tight_layout()
plt.show()
```
