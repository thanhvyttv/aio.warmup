```python
import numpy as np
import matplotlib.pyplot as plt

#  Tính đạo hàm bằng sai phân hữu hạn
def numerical_derivative(f, x, h=1e-5):
    return (f(x+h) - f(x)) / h

# Hàm vẽ đồ thị của hàm số và đạo hàm
def plot_chain_rule(f, g, x_range=(-2, 2), title = "Đạo hàm Chain Rule"):
    x_vals = np.linspace(x_range[0], x_range[1], 400)

    #Tính giá trị của g(x) và f(g(x))
    g_vals = g(x_vals)
    f_vals = f(g_vals)

    #Tính đạo hàm theo chain rule
    chain_rule_derivative_vals = numerical_derivative(f, g_vals) * numerical_derivative(g, x_vals)

    # Vẽ đồ thị của f(g(x)) và đạo hàm
    plt.figure(figsize=(8,5))
    plt.plot(x_vals, f_vals, label = 'f(g(x))', color = 'blue')
    plt.plot(x_vals, chain_rule_derivative_vals, label = "f'(x) theo chain rule", color = 'red', linestyle = '--')

    plt.axhline(0, color='black', linewidth = 0.5, linestyle = '--')
    plt.axvline(0, color='black', linewidth = 0.5, linestyle = '--')

    plt.xlabel('x')
    plt.ylabel('Value')
    plt.title(title)
    plt.legend()
    plt.grid()
```

```python
def f1(x):
    return np.exp(x)
def g1(x):
    return x**3 - 2*x + 1

plot_chain_rule(f1, g1)
```

```python
def f2(x):
    return np.log(x)

def g2(x):
    return 5*x**2 + 3*x + 1

plot_chain_rule(f2, g2, (-1, 2)) 
```

```python
def f3(x):
    return np.sqrt(x)

def g3(x):
    return x**4 + 2*x**2 + 1

plot_chain_rule(f3, g3)
```
