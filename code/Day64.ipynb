```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Bi 1: c v hin th nh
img_path = "data/ComputerVidion/image.jpg"
img = cv2.imread(img_path, 1)

converted_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(converted_image)
plt.axis("off")
plt.show()
```

```python
cv2.imshow("window name", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Bi 2: Lu nh di tn khc
cv2.imwrite("data/ComputerVidion/saved_image.jpg", img)
```

```python
# Bi 3: Chuyn nh t BGR sang RGB
converted_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(converted_image)
plt.axis("off")
plt.show()
```

```python
#  Bi 4: Chuyn nh t BGR sang HSV
converted_img_bgr_to_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(converted_img_bgr_to_hsv)
plt.axis("off")
plt.show()
```

```python
# Bi 5: Chuyn nh sang Grayscale
converted_img_2_gray_scale = cv2.cvtColor(converted_image, cv2.COLOR_BGR2GRAY)
plt.imshow(converted_img_2_gray_scale)
plt.axis("off")
plt.show()
```

```python
# Bi 6: So snh nh RGB v HSV
combined = cv2.hconcat([converted_image, converted_img_bgr_to_hsv])
plt.imshow(combined)
plt.axis('off')
plt.show()
```
