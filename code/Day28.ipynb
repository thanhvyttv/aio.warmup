```python
import numpy as np
import matplotlib.pyplot as plt

# Hàm tính đạo hàm bằng sai phân hữu hạn
def numerical_derivative(f, x, h=1e-5):
    return (f(x + h) -  f(x))/h

# Hàm vẽ đồ thị
def plot_function_and_derivative(f, x_range=(-2, 2), title="Đồ thị f(x) và f'(x)"):
    x_vals = np.linspace(x_range[0], x_range[1], 400)
    y_vals = f(x_vals)
    y_derivative_vals = numerical_derivative(f, x_vals)

    plt.figure(figsize=(8, 5))
    plt.plot(x_vals, y_vals, label="f(x)", color="blue")
    plt.plot(x_vals, y_derivative_vals, label="f'(x)", color="red", linestyle ="--")
    plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
    plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
    plt.xlabel("x")
    plt.ylabel("f(x) & f'(x)")
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.savefig(f'{title}.png')  # Save the plot before showing it
    plt.show()
```

```python
def f1(x):
    return x**2 - 3*x + 2

plot_function_and_derivative(f1, (-2, 4), "Bài 1")
```

```python
def f2(x):
    return x**3 - 3*x**2 + 4*x - 2

plot_function_and_derivative(f2, (-2, 4), "Bài 2")
```

```python
def f3(x):
    return np.exp(x)

plot_function_and_derivative(f3, (-2,4), "Bài 3")
```

```python
def f4(x):
    return np.log(x)

plot_function_and_derivative(f4, (0.1, 4), "Bài 4")
```

```python
def f5(x):
    return np.sin(x)

plot_function_and_derivative(f5, (-2*np.pi,2*np.pi), "Bài 5")
```

```python
def f6(x):
    return np.cos(x)

plot_function_and_derivative(f6, (-2*np.pi, 2*np.pi), "Bài 6")
```

```python
def f7(x):
    return np.tan(x)
plot_function_and_derivative(f7, (-np.pi/2 + 0.1, np.pi/2 - 0.1), "Bài 7")
```

```python
def f8(x):
    return np.sqrt(x)

plot_function_and_derivative(f8, (0,4), "Bài 8")
```
